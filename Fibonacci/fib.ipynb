Plotting the execution time of this iterative Fibonacci algorithm

```python
import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pylab as pl
```

```python
def fib(n):
    
    l = []
    c = []
    start_time = time.time()
    for i in range(n):
        count = 0
        x = 0
        y = 1
        count+=2
        for _ in range(i):
            y = x + y
            x = y - x
            count+= 2
        count+=1
            
        l.append(time.time() - start_time)
        c.append(count)
    
    return l, c
```

```python

l , i = fib(10000)
y = np.array(l)
z = np.array(i)



plt.plot(y)
plt.xlim(0,10000)
plt.ylim(0,11)    
plt.xlabel('Nth term')
plt.ylabel('time (s)')
plt.title('Iterative Fibonacci Execution time')

plt.show()
pl.close()



plt.plot(z)
plt.xlim(0,10000)

    
plt.xlabel('Nth term')
plt.ylabel('Number of instrucions')
plt.title('Iterative Fibonacci Instruction count');
```
